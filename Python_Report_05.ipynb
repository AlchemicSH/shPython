다형성이란 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 기능을 말합니다. 제가 여기서 제시하고자 하는 저만의 다형성 코드는 다음과 같습니다.</font>

```python
class Person :
    name = ''
    age = 0
    sex = ''

    def __init__(self, name, age, sex) :
        self.name = name
        self.age = age
        self.sex = sex

    def printThis(self) :
        print 'Name : ' + self.name + ', Age : ' + str(self.age) + ', Sex : ' + self.sex
        
class Employee(Person) :
    position = ''
    month_salary = 0

    def __init__(self, name, age, sex, position, month_salary) :
        Person.__init__(self, name, age, sex)
        self.position = position
        self.month_salary = month_salary

    def printThis(self) :
        print 'Name : ' + self.name + ', Age : ' + str(self.age) + ', Sex : ' + self.sex
        print 'Position : ' + self.position + ', Salary : ' + str(self.month_salary)

class Student(Person) :
    school = ''
    grade = ''

    def __init__(self, name, age, sex, school, grade) :
        Person.__init__(self, name, age, sex)
        self.school = school
        self.grade = grade

    def printThis(self) :
        print 'Name : ' + self.name + ', Age : ' + str(self.age) + ', Sex : ' + self.sex
        print 'School : ' + self.school + ', Grade : ' + self.grade

p = Person('Mike', 40, 'Male')
e1 = Employee('Jhon', 32, 'Male', 'SalsePerson', 300)
e2 = Employee('Lisa', 25, 'Female', 'Temporary', 125)
s = Student('Jesica', 18, 'Female', 'High School', 'Second grade')

p.printThis()
print
e1.printThis()
print
e2.printThis()
print
s.printThis()
```

문제 3번은은 문제에서 요구하는 모든 사항을 만족하는 Counter 클래스를 작성하는 것입니다.
<br>
<br>
소스 코드는 다음과 같습니다.</font>

```python
class Counter :
    def __init__(self, count, step = 1) :    # 생성자
        self.count = count
        self.step = step
        
    def __str__(self) :   # 인스턴스 자체를 출력할 수 있는 메소드
        temp = "[Count(step:" + str(self.step) + ")] " + str(self.count)
        return temp
    
    def incr(self) :     # step에 주어진 증분만큼 count를 증가시키는 메소드
        self.count = self.count + self.step
        
    def __call__(self) :   # 인스턴스 객체를 직접 호출할 수 있는 메소드
        self.count = self.count + self.step
        
    def __add__(self, add_value) :   # + 연산자 오버로딩
        self.count = self.count + add_value
        return self
    
    def __sub__(self, sub_value) :   # - 연산자 오버로딩
        self.count = self.count - sub_value
        return self
    
    def __cmp__(self, value) :    # 비교 연산자 오버로딩
        return self.count - value


c = Counter(10)
d = Counter(10, 2)

print c
print d
print

c.incr()
d.incr()
print c
print d
print

c()
d()
print c
print d
print

c = c + 5
d = d - 5
print c
print d
print

print c > 10
print d > 10
print c < 10
print d < 10
print c == 17
print d != 9
```

문제 4번은 이미 정의되어 있는 MySet 클래스를 다른 사람에게 가르친다고 생각하며 설명을 하는 것입니다.</font>

```python
class MySet(list):
    def __init__(self, l):    # 생성자
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):    # 출력 함수 오버로딩
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):    # 정적 메소드 (리스트 내의 원소 중복 제거)
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)


if __name__ == "__main__":    # 메인에서 실행하도록 구성되어짐.
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

<font size = 3pt>
먼저 MySet 클래스는 인스턴스 객체를 생성하게 되면 리스트 형태로 생성이 되고 초기화가 됩니다. 먼저, 생성자에 의해서 self에 입력받은 리스트의 원소를 넣어둔 뒤, 정적 메소드인 eliminate_duplicate() 메소드를 호출하여 self 내의 중복 원소를 제거합니다. 그 다음 출력을 하게 되면 출력 함수 오버로딩 메소드를 통해 코드 작성자가 따로 정의한 대로 출력을 하게 됩니다. 생성자에서 자기 자신의 정적 메소드를 호출하여 사용을 할 수 있는 이유는 정적 메소드는 원래 인스턴스 객체 없이 클래스에서 직접 호출이 가능하기 때문입니다. 중복 원소를 제거하는 상황을 살펴보면 자신의 인스턴스 내에서 정적 메소드를 호출하여 사요을 하는 것이기 때문에 정적 메소드로 self를 넘겨주게 되면 self 자체를 수정을 하게 됩니다. 따라서 따로 return self 를 하지 않아도 중복 원소를 쉽게 제거할 수 있게 되는 것입니다.

문제 5번은 4번 문제에 정의된 MySet 클래스에 문제에서 요구하는 메소드를 더 추가하여 문제에서 요구사항을 만족하는 코드를 작성하는 것입니다.
<br>
소스코드는 다음과 같습니다.</font>

```python
class MySet(list):
    def __init__(self, l):   # 생성자
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):    # 출력 함수 오버로딩
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    def __or__(self, other) :   # |(or) 연산자 오버로딩(두 집합의 합집합 반환)
        self = self + other
        MySet.eliminate_duplicate(self) # 정적 메소드 호출
        return MySet(self)

    def __and__(self, other) :   # & 연산자 오버로딩(두 집합의 교집합 반환)
        s = []
        for e in self :
            if e in other :
                s.append(e)
        self[:] = []
        for e in s :
            self.append(e)

        return MySet(self)

    def __sub__(self, other) :    # - 연산자 오버로딩(두 집합의 차집합 반환)
        s = []
        for e in self :
            if e not in other :
                s.append(e)
        self[:] = []
        for e in s :
            self.append(e)

        return MySet(self)
    
    @staticmethod
    def eliminate_duplicate(l):   # 정적 메소드(리스트 내의 중복 원소 제거)
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
    print
    s = s | t
    print s
    print
    s1 = MySet([1,2,3])
    t1 = MySet([3,4,5])
    s1 = s1 & t1
    print s1
    print
    s2 = MySet([1,2,3])
    t2 = MySet([3,4,5])
    s2 = s2 - t2
    print s2
```

```python
import urllib
import string
import operator

class WebWordsFrequency :
    urls = []
    words_dic = {}
    max_frequency_string_list = []

    def __init__(self, *url) :
        for temp in url :
            self.urls.append(temp)

    def addUrl(self, url) :
        self.urls.append(url)

    def removeUrl(self, url) :
        self.urls.remove(url)

    def listUrls(self) :
        for url in self.urls :
            print url

    def getWordsFrequency(self) :
        for url in self.urls :
            s = urllib.urlopen(url).read()

            s.lower()

            while(True) :
                start = -1
                finish = -1

                start = s.find('<style')
                finish = s.find('</style')

                if(start == -1 and finish == -1) :
                    break
                else :
                    s = s[:start] + ' ' + s[finish+8:]

            while(True) :
                start = -1
                finish = -1

                start = s.find('<script')
                finish = s.find('</script')

                if(start == -1 and finish == -1) :
                    break
                else :
                    s = s[:start] + ' ' + s[finish+9:]

            while(True) :
                start = -1
                finish = -1

                start = s.find('<')
                finish = s.find('>')

                if(start < finish) :
                    s = s[:start] + ' ' + s[finish+1:]
                elif(finish != -1) :
                    s = s[:finish] + ' ' + s[finish+1:]
                else :
                    break

            words = s.split()

            for word in words :
                word = word.translate(string.maketrans('',''), string.punctuation).strip()

                if word != '' :
                    if word in self.words_dic :
                        self.words_dic[word] += 1
                    else :
                        self.words_dic[word] = 1
        
        for key in self.words_dic :
            print key, self.words_dic[key]

    def getMaxFrequencyWords(self) :
        max_frequency = -999

        for key in self.words_dic :
            if(max_frequency < self.words_dic[key]) :
                max_frequency = self.words_dic[key]
                self.max_frequency_string_list = []
                self.max_frequency_string_list.append(key)
            elif(max_frequency == self.words_dic[key]) :
                self.max_frequency_string_list.append(key)

        for string in self.max_frequency_string_list :
            print string

class OrderedWebWordsFrequency(WebWordsFrequency) :
    words_dic = {}
    
    def __init__(self, *url) :
        WebWordsFrequency.__init__(self, *url)

    def getWordsFrequency(self) :
        for url in self.urls :
            s = urllib.urlopen(url).read()

            s.lower()

            while(True) :
                start = -1
                finish = -1

                start = s.find('<style')
                finish = s.find('</style')

                if(start != -1 and finish != -1 and start < finish) :
                    s = s[:start] + ' ' + s[finish+8:]
                else :
                    break

            while(True) :
                start = -1
                finish = -1

                start = s.find('<script')
                finish = s.find('</script')

                if(start != -1 and finish != -1 and start < finish) :
                    s = s[:start] + ' ' + s[finish+9:]
                else :
                    break

            while(True) :
                start = -1
                finish = -1

                start = s.find('<')
                finish = s.find('>')

                if(start != -1 and finish != -1 and start < finish) :
                    s = s[:start] + ' ' + s[finish+1:]
                elif(finish != -1) :
                    s = s[:finish] + ' ' + s[finish+1:]
                else :
                    break

            words = s.split()

            for word in words :
                word = word.translate(string.maketrans('',''), string.punctuation).strip()

                if word != '' :
                    if word in self.words_dic :
                        self.words_dic[word] += 1
                    else :
                        self.words_dic[word] = 1

        sorted_words_list = sorted(self.words_dic.iteritems(), key = operator.itemgetter(1), reverse = True)

        for word, frequency in sorted_words_list :
            print word, frequency
        
w1 = WebWordsFrequency('http://www.daum.net')
w1.addUrl('http://www.naver.com')
w1.addUrl('http://www.nate.com')
w1.listUrls()
print
w1.getWordsFrequency()
print
print '***************************************************'
w1.getMaxFrequencyWords()
print '***************************************************'
print
print
print
w2 = OrderedWebWordsFrequency('http://www.naver.com', 'http://www.daum.net')
w2.getWordsFrequency()
```
